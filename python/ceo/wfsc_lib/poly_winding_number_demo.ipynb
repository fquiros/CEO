```python
import numpy as np
import matplotlib.pyplot as plt
from ceo.wfsc_lib.poly_winding_number import poly_winding_number
```

# Polygon winding number
Test wether a point is enclosed by a polygon.

## Define the vertices of a polygon

```python
poly = np.zeros((5,2))
poly[:,0] = [ 0, 10,   0, -10,  0]
poly[:,1] = [10,  0, -10,   0, 10]

plt.plot(poly[:,0], poly[:,1]);
```

## Define an array that encloses the polygon area

```python
vec1 = np.arange(-15,15)
nPx = len(vec1)
x2d, y2d = np.meshgrid(vec1,vec1)
pnts = np.vstack((x2d.ravel(), y2d.ravel())).T
pnts.shape
```

```python
plt.plot(poly[:,0], poly[:,1])
plt.plot(pnts[:,0], pnts[:,1], '+')
```

## Find mask to points enclosed by the polygon
Points at the border are also selected.

```python
w = poly_winding_number(pnts, poly)
```

```python
plt.imshow(w.reshape((nPx,nPx)))
plt.colorbar()
```

### Return points coordinates enclosed by the polygon
You could also retrieve the coordinate vectors by setting the keyword **return_points**.

```python
w, encl_pnts = poly_winding_number(pnts, poly, return_points=True)
```

```python
encl_pnts.shape
```
